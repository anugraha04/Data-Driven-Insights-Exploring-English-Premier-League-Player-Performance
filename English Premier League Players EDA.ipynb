# Performance Trends & Data Insights: English Premier League Players


This analysis explores the performance of English Premier League players, focusing on goal contribution, defensive efficiency, and player trends.

The dataset is correct upto and including 2020-09-24.

# Data Collection & Understanding

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns 
```

```python
epl=pd.read_csv('EPL PLayers.csv')
```

## Analysis w.r.t Player Nationality

### Top 10 Nationalities by Player Count

```python
plt.figure(figsize=(8,5))

top_nationalities = epl['Nationality'].value_counts().head(10).reset_index()
top_nationalities.columns = ['Nationality', 'Count']  

sns.barplot(x='Nationality', y='Count', data=top_nationalities, palette="pastel")

plt.title('Top 10 Nationalities by Player Count')
plt.xticks(rotation=90)

plt.show()
```

### Top 10 Nationalities by Appearance

```python
plt.figure(figsize=(8,7))
appearance_grouped = epl[['Appearances', 'Nationality']].groupby('Nationality').sum().sort_values(by='Appearances', ascending=False).head(10).reset_index()
plt.ylabel('Top 10 Nationality by Appearance')
sns.barplot(x='Appearances',y='Nationality',data=appearance_grouped, palette="pastel")
plt.show()
```

### Goal Distribution by Nationality

```python
epl_sorted = epl.sort_values(by='Goals', ascending=False)

top_n = 10
top_players = epl_sorted.head(top_n)

others_goals = epl_sorted.iloc[top_n:]['Goals'].mean()
others_row = pd.DataFrame({'Goals': [others_goals], 'Nationality': ['Others']})

epl_grouped = pd.concat([top_players, others_row])

epl_grouped = epl_grouped.groupby('Nationality', as_index=False).sum()

plt.figure(figsize=(8, 8))
plt.title('Player of which country scores the most goals')
plt.pie(epl_grouped['Goals'], labels=epl_grouped['Nationality'], autopct='%0.1f%%', colors=sns.color_palette("pastel", len(epl_grouped))  
)
plt.show()
```

### Top 10 Nationalities by Total Wins

```python
plt.figure(figsize=(8,7))

wins_grouped = epl[['Wins', 'Nationality']].groupby('Nationality').sum().sort_values(by='Wins', ascending=False).head(10).reset_index()

sns.barplot(x='Wins', y='Nationality', data=wins_grouped, palette="pastel")  

plt.xlabel('Number of Wins')
plt.ylabel('Nationality')
plt.title('Top 10 Nationalities by Total Wins')

plt.show()
```

## Feature Engineering

### Player statistics: Top 5 Goalkeepers 

```python
epl['Clean sheets per match']=epl['Clean sheets']/epl['Appearances']
epl['Saves per match']=epl['Saves']/epl['Appearances']
epl['High Claims per match']=epl['High Claims']/epl['Appearances']
epl['Catches per match']=epl['Catches']/epl['Appearances']

player_clean = epl[['Name', 'Club', 'Clean sheets']].sort_values(by='Clean sheets', ascending=False).head()
player_clean_pm = epl[['Name', 'Club', 'Clean sheets per match']].sort_values(by='Clean sheets per match', ascending=False).head()

player_saves = epl[['Name', 'Club', 'Saves']].sort_values(by='Saves', ascending=False).head()
player_saves_pm = epl[['Name', 'Club', 'Saves per match']].sort_values(by='Saves per match', ascending=False).head()

player_hc = epl[['Name', 'Club', 'High Claims']].sort_values(by='High Claims', ascending=False).head()  
player_hc_pm = epl[['Name', 'Club', 'High Claims per match']].sort_values(by='High Claims per match', ascending=False).head()

player_catches = epl[['Name', 'Club', 'Catches']].sort_values(by='Catches', ascending=False).head()
player_catches_pm = epl[['Name', 'Club', 'Catches per match']].sort_values(by='Catches per match', ascending=False).head()

fig, axes = plt.subplots(4, 2, figsize=(14, 10))

metrics = [
    ('Clean sheets', player_clean,'Clean sheets', axes[0, 0]),
    ('Clean sheets per match', player_clean_pm,'Clean sheets per match', axes[0, 1]),
    ('Saves', player_saves,'Saves', axes[1, 0]),
    ('Saves per match', player_saves_pm,'Saves per match', axes[1, 1]),
    ('High Claims', player_hc,'High Claims', axes[2, 0]),
    ('High Claims per match', player_hc_pm,'High Claims per match', axes[2, 1]),
    ('Catches', player_catches,'Catches', axes[3, 0]),
    ('Catches per match', player_catches_pm,'Catches per match', axes[3, 1])
]

for title, data, col, ax in metrics:
    sns.barplot(y='Name', x=col, hue='Club', data=data, ax=ax, palette='pastel')
    ax.set_title(title)
    ax.set_xlabel('Count')
    ax.legend(title='Club', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

### Top 5 Defenders statistics

```python
epl['Blocked shots per match']=epl['Blocked shots']/epl['Appearances']
epl['Interceptions per match']=epl['Interceptions']/epl['Appearances']
epl['Clearances per match']=epl['Clearances']/epl['Appearances']
epl['Headed Clearance per match']=epl['Headed Clearance']/epl['Appearances']
epl['Clearances off line per match']=epl['Clearances off line']/epl['Appearances']

player_bs = epl[['Name', 'Club', 'Blocked shots']].sort_values(by='Blocked shots', ascending=False).head()
player_bs_pm = epl[['Name', 'Club', 'Blocked shots per match']].sort_values(by='Blocked shots per match', ascending=False).head()

player_in = epl[['Name', 'Club', 'Interceptions']].sort_values(by='Interceptions', ascending=False).head()
player_in_pm = epl[['Name', 'Club', 'Interceptions per match']].sort_values(by='Interceptions per match', ascending=False).head()

player_cl = epl[['Name', 'Club', 'Clearances']].sort_values(by='Clearances', ascending=False).head()  
player_cl_pm = epl[['Name', 'Club', 'Clearances per match']].sort_values(by='Clearances per match', ascending=False).head()  

player_hcl = epl[['Name', 'Club', 'Headed Clearance']].sort_values(by='Headed Clearance', ascending=False).head()
player_hcl_pm = epl[['Name', 'Club', 'Headed Clearance per match']].sort_values(by='Headed Clearance per match', ascending=False).head()

player_cloff = epl[['Name', 'Club', 'Clearances off line']].sort_values(by='Clearances off line', ascending=False).head()
player_cloff_pm = epl[['Name', 'Club', 'Clearances off line per match']].sort_values(by='Clearances off line per match', ascending=False).head()


fig, axes = plt.subplots(5, 2, figsize=(14, 12)) 

metrics = [
    ('Blocked shots', player_bs,'Blocked shots', axes[0, 0]),
    ('Blocked shots per match', player_bs_pm,'Blocked shots per match', axes[0, 1]),
    ('Interceptions', player_in,'Interceptions', axes[1, 0]),
    ('Interceptions per match', player_in_pm,'Interceptions per match', axes[1, 1]),
    ('Clearances', player_cl,'Clearances', axes[2, 0]),
    ('Clearances per match', player_cl_pm,'Clearances per match', axes[2, 1]),
    ('Headed Clearance', player_hcl,'Headed Clearance', axes[3, 0]),
    ('Headed Clearance per match', player_hcl_pm,'Headed Clearance per match', axes[3, 1]),
    ('Clearances off line', player_cloff,'Clearances off line', axes[4, 0]),
    ('Clearances off line per match', player_cloff_pm,'Clearances off line per match', axes[4, 1])
]

for title, data, col, ax in metrics:
    sns.barplot(y='Name', x=col, hue='Club', data=data, ax=ax, palette="pastel")
    ax.set_title(title)
    ax.set_xlabel('Count')
    ax.legend(title='Club', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

# Insights and Key Takeaways


1. The most represented foreign nations (1st France, 2nd Spain and 3rd Brazil) have maintained their ranks through parameters like top Appearances and Total wins. 
2. Argentina, despite having comparatively lower player count has managed to score the 2nd most Goals.
3. Manchester-City has scored the most Goals, however, Liverpool has the highest Total Wins. 
4. Liverpool (most Wins) has the most balanced Appearance ratio of Goalkeepers, Defenders, Midfielders and Forwards compared to other Clubs. 
5. Goalkeepers and Defender are relatively older than thier mid-field and attacking colleagues. (Fact: Liverpool is the oldest group of players.)
6. Majority of players come from Defence (34%) and Midfield position (34.7%). Usually midfielders are versatile and can play in defence or in attack if needed validating their importance in the field.
7. Some players (for example, Jordan Henderson) have made it to Top 5 Wins despite lower Appearances, proving their skills in most of their appearances.
8. Joe Hart from Tottenham-Hotspur has exceptionally excelled in almost every aspect as a Goalkeeper.
9. Phil Jagielka from Shffield-United has the most Interceptions, Clearances and Clearnces off line, being one of the best Denfenders and midfielders.
10. Sergio Aguero from Manchester-City has been the top Goal contributor till date.
11. Goals summary:
Most goals: Sergio Aguero, Harry Kane, Jamie Vardy
Most right foot goals: Sergio Aguero, Harry Kane, Jamie Vardy
Most left foot goals: Mohammed Salah, Oliver Giroud, Ryhad Mahrez
Most headed goals: Oliver Giroud, Christian Benket, Andy Carrol
Most goal scoring nations: England, France, Brazil Argentina
Most goal scoring clubs: Machester City, Liverpool and Tottenham
Fewest goals: Leeds United
12. Surprisingly Sergio Aguero (who's the top scorer) is also a big chance squanderer.
13. Joe Hart (one of the best goalkeepers)is also the player with most Errors leading to goal.
14. Phil Jagielka apart from excelling as a Defender also has the most number of Own goals.
15. Mark Noble from West-Ham-United has the highest number Red cards among all the 571 players.
